```python
#clean the data
df["Id"] = df["Id"].astype(str)
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"], format= "%m/%d/%Y")
df.dtypes
```

```python
df["distance_diff"] = df["TotalDistance"] - df["TrackerDistance"]
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.rename(columns = { 'activitydate': 'activity_date', 'totalsteps': 'total_steps', 'totaldistance':'total_distance', 'trackerdistance':'trackerd_istance',
       'loggedactivitiesdistance':'logged_activities_distance', 'veryactivedistance':'very_active_distance',
       'moderatelyactivedistance':'moderately_active_distance', 'lightactivedistance':'light_active_distance',
       'sedentaryactivedistance':'sedentary_active_distance', 'veryactiveminutes':'very_active_minutes', 'fairlyactiveminutes':'fairly_active_minutes',
       'lightlyactiveminutes':'lightly_active_minutes', 'sedentaryminutes':'sedentary_minutes'}, inplace=True)
```

```python
#create Column
day_of_week = df["activity_date"].dt.day_name()
df["day_of_week"]= day_of_week
df["n_day_of_week"]= df["activity_date"].dt.weekday  #0 monday 6 sunday
```

```python
#subset the data
df = df[['id', 'activity_date', 'total_steps', 'total_distance',
       'trackerd_istance', 'logged_activities_distance',
       'very_active_distance', 'moderately_active_distance',
       'light_active_distance', 'sedentary_active_distance',
       'very_active_minutes', 'fairly_active_minutes',
       'lightly_active_minutes', 'sedentary_minutes', 'calories',
       'distance_diff', 'day_of_week', 'n_day_of_week'
       ]].copy()
```

```python
df.head(4)
```

```python
#Sum of all active minutes excluding sedentary minutes
df['total_active_minutes'] = df['very_active_minutes'] + df['fairly_active_minutes'] 
df
```

```python
#Analysis
#Category-1
#sedentary: less than 6000 on avg
#active: b/w 6000 to 12000 on avg
#very active: more than 12000 on avg

df["id"].unique()        #Total 33 peapole are tracked
id_grp = df.groupby(["id"])
id_avg_step = id_grp["total_steps"].mean().sort_values(ascending=False)
id_avg_step = id_avg_step.to_frame()

condition = [(id_avg_step <= 6000), (id_avg_step > 6000) &(id_avg_step <12000) , (id_avg_step >= 12000)]
values = ["sedentary", "active", "very_active"]
id_avg_step["activity_level"] = np.select(condition,values)

id_activity_level = id_avg_step["activity_level"]
id_avg_step

df["activity_level"] = [id_activity_level[c] for c in df["id"]]

```

```python
#Category-2 (all activity include very acive, active, low active and , sedentary active)
#low use: less than 8 hours per day
#normal use: b/w 8 and 16 hrs per day
#high use: more than 16 hrs 
df["id"].unique()        #Total 33 peapole are tracked
id_grp = df.groupby(["id"])
id_time = id_grp["total_active_minutes"].sum().sort_values(ascending=False)
id_time = id_time.to_frame()

condition = [(id_time <= 480), (id_time > 480) &(id_time <960) , (id_time > 960)]
values = ["low use", "normal use", "high use"]
id_time["use category"] = np.select(condition,values)

id_use_level = id_time["use category"]

df["use category"] = [id_use_level[c] for c in df["id"]]
```

```python
#correlation b/w Total activity time and calories burned
#we excluded sedentary time because we don't want any output from that

axis = sns.scatterplot(x= "total_active_minutes", y="calories", data = df, hue=df["use category"])

plt.title("Correlation calories vs. Total_activity_time")

plt.tight_layout()
plt.show
```

```python
#correlation b/w steps and calories burned

axis = sns.scatterplot(x= "total_steps", y="calories", data = df, hue=df["activity_level"])

plt.title("Correlation calories vs. Steps")

plt.tight_layout()
plt.show
```

```python
#Average steps per day
fig, ax = plt.subplots(1,1, figsize=(8,5))

day_grp = df.groupby(["day_of_week"])
avg_daily_steps = day_grp["total_steps"].mean()
avg_steps = df["total_steps"].mean()

plt.bar(avg_daily_steps.index, avg_daily_steps, color="green")

ax.axhline(y=avg_daily_steps.mean(), color = "green", label= "avg_daily_steps")

ax.set_ylabel("Steps")
ax.set_xlabel("Day of week")
ax.set_title("Avg number of steps per day")
plt.legend()
plt.show()

```

```python
#Percentage of activity in minutes
very_active_mins = df["very_active_minutes"].sum()
fairly_active_mins =df["fairly_active_minutes"].sum()
ligtly_active_mins = df["lightly_active_minutes"].sum()
sedentary_mins = df["sedentary_minutes"].sum()
slice_colors = ['blue', 'green', 'red', 'purple']

slices = [very_active_mins,fairly_active_mins,ligtly_active_mins, sedentary_mins ]
labels = ["very_active_mins","fairly_active_mins","ligtly_active_mins", "sedentary_mins"]
explode = [0,0,0,0.1]
  
plt.pie(slices,labels=labels, explode=explode, autopct = "%1.1f%%", colors=slice_colors)
plt.title("% of activities level in minutes")


```

```python
#Correlation between activity level in minutes and calories
n_day_of_week = [0,1,2,3,4,5,6]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize= (8,9), dpi=70)

sns.scatterplot(data=df, x= "calories", y= "sedentary_minutes", hue="activity_level", ax=axes[0,0],legend=False)
sns.scatterplot(data=df, x= "calories", y= "lightly_active_minutes", hue="activity_level", ax=axes[0,1],legend=False)
sns.scatterplot(data=df, x= "calories", y= "fairly_active_minutes", hue="activity_level", ax=axes[1,0],legend=False)
sns.scatterplot(data=df, x= "calories", y= "very_active_minutes", hue="activity_level", ax=axes[1,1],legend=True)

plt.legend(title ="Activity level") 
plt.show()
```
